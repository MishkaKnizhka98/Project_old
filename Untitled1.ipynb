```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
data = pd.read_csv("student-mat.csv")
```

```python
school_name = input("Choose the school: ")
data = data[data["school"] == school_name]

data = data[["sex", "age", "famsize", "Pstatus", "Mjob", "Fjob", "higher", "activities", "G3"]]
data.dtypes
```

```python
data["Mjob"] = data["Mjob"].apply(lambda x: "m_" + x)
data["Fjob"] = data["Fjob"].apply(lambda x: "f_" + x)
```

```python
data["Mjob"]
```

```python
pd.get_dummies(data["Mjob"])
pd.get_dummies(data["Fjob"])
```

```python
dummies = pd.concat([pd.get_dummies(data["Mjob"]), pd.get_dummies(data["Fjob"])], axis = 1) 
dummies
```

```python
data = pd.concat([data, dummies], axis = 1)
```

```python
data = data.drop(["Mjob", "Fjob"], axis = 1)
data
```

```python
non_num = data.select_dtypes(include = "object")
for column in non_num.columns:
    print(column, ": ", non_num[column].unique())
```

```python
encoder = LabelEncoder()

for column in non_num.columns:
    data[column] = encoder.fit_transform(data[column])
```
